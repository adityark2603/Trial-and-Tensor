iris Dataset - Train/Validation/Test Split


```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Load Iris dataset
iris = datasets.load_iris()

df = pd.DataFrame(iris.data)
label = pd.DataFrame(iris.target)

# Split 70% Train+Val and 30% Test
X_train, X_test, y_train, y_test = train_test_split(
    df, label, test_size=0.3, shuffle=True
)

# Split 70% into Train and Validation (70:30)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.3, shuffle=True
)

print("Total samples:", df.shape[0])
print("Training samples:", X_train.shape[0])
print("Validation samples:", X_val.shape[0])
print("Testing samples:", X_test.shape[0])

print("\nFirst 3 samples of Training set")
print(X_train.head(3))

print("\nRandom 3 samples of Training set")
print(X_train.sample(3))

print("\nFirst 3 samples of Validation set")
print(X_val.head(3))

print("\nRandom 3 samples of Validation set")
print(X_val.sample(3))

print("\nFirst 3 samples of Testing set")
print(X_test.head(3))

print("\nRandom 3 samples of Testing set")
print(X_test.sample(3))

```

Digits Dataset – Train / Validation / Test Split

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Load Digits dataset
digits = datasets.load_digits()

df = pd.DataFrame(digits.data)
label = pd.DataFrame(digits.target)

# Split 80% Train+Val and 20% Test
X_train, X_test, y_train, y_test = train_test_split(
    df, label, test_size=0.2, shuffle=True
)

# Split remaining 80% into Train and Validation (80:20)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, shuffle=True
)

print("Total samples:", df.shape[0])
print("Training samples:", X_train.shape[0])
print("Validation samples:", X_val.shape[0])
print("Testing samples:", X_test.shape[0])

print("\nFirst 3 samples of Training set")
print(X_train.head(3))

print("\nRandom 3 samples of Training set")
print(X_train.sample(3))

print("\nFirst 3 samples of Validation set")
print(X_val.head(3))

print("\nRandom 3 samples of Validation set")
print(X_val.sample(3))

print("\nFirst 3 samples of Testing set")
print(X_test.head(3))

print("\nRandom 3 samples of Testing set")
print(X_test.sample(3))
```

Diabetes Dataset – Train / Validation / Test Split

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Load Diabetes dataset
diabetes = datasets.load_diabetes()

df = pd.DataFrame(diabetes.data)
label = pd.DataFrame(diabetes.target)

# Split 75% Train+Val and 25% Test
X_train, X_test, y_train, y_test = train_test_split(
    df, label, test_size=0.25, shuffle=True
)

# Split remaining 75% into Train and Validation (70:30)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.3, shuffle=True
)

print("Total samples:", df.shape[0])
print("Training samples:", X_train.shape[0])
print("Validation samples:", X_val.shape[0])
print("Testing samples:", X_test.shape[0])
```
