```python
# Experiment-1(a): Access and Observe Iris Dataset
# Duration: 4 Hours

# Import required libraries
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split

print('THIS IS THE PROGRAM TO ACCESS IRIS DATASET\n')

# Load Iris dataset
iris = datasets.load_iris()

# Print dataset description
print("To print the description of Iris Dataset\n")
print(iris.DESCR)
print("\n" + "="*80 + "\n")

# Convert dataset into DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add target labels
df['target'] = pd.Series(iris.target)

# Add target names
df['target_names'] = df['target'].apply(lambda y: iris.target_names[y])

# Display first 5 samples
print("To display first 5 samples of the dataset:\n")
print(df.head(5))

print("\n" + "="*80 + "\n")

# Display randomly selected 5 samples
print("To display randomly selected 5 samples from the dataset:\n")
print(df.sample(5))

print("\n" + "="*80 + "\n")

# Train-Test Split (70:30)
df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)

# Display dataset sizes
print("Dataset Split Information:\n")
print("Total number of samples in the dataset =", df.shape[0])
print("Number of samples in training set =", df_train.shape[0])
print("Number of samples in testing set =", df_test.shape[0])

print("\n" + "="*80 + "\n")

# Display first five samples of training set
print("First five samples of Training Set:\n")
print(df_train.head(5))

print("\n" + "="*80 + "\n")

# Display first five samples of testing set
print("First five samples of Testing Set:\n")
print(df_test.head(5))
```
